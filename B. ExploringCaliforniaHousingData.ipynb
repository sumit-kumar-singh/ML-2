```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# getting housing data
housing_data = pd.read_csv('datasets/housing.csv')
# getting top 5 records of data
housing_data.head()
```

```python
# getting random sample of data
housing_data.sample(5)
```

```python
# getting shape of dataframe
print('Rows and Columns :', housing_data.shape)
```

```python
# data cleaning process
housing_data = housing_data.dropna()
```

```python
# getting shape of dataframe after cleaning
print('Rows and Columns :', housing_data.shape)
```

```python
# getting statistic information about the dataset
housing_data.describe()
```

```python
# getting unique values from particular column
# here ocean_proximity is categorical data
housing_data['ocean_proximity'].unique()
```

```python
# plotting scatter plot total rooms vs median house price
fix, ax = plt.subplots(figsize=(12, 8))
plt.scatter(housing_data['total_rooms'], housing_data['median_house_value'])
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')
```

```python
# plotting scatter plot median housing age vs median house price
fix, ax = plt.subplots(figsize=(12, 8))
plt.scatter(housing_data['housing_median_age'], housing_data['median_house_value'])
plt.xlabel('Housing Median Age')
plt.ylabel('Median House Value')
```

```python
# plotting scatter plot median income vs median house price
fix, ax = plt.subplots(figsize=(12, 8))
plt.scatter(housing_data['median_income'] * 1000, housing_data['median_house_value'])
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
```

```python
# getting co-realtion in dataset
# +ve corelation and -ve corelation (-1 to +1)
housing_data_corr = housing_data.corr()
housing_data_corr
```

```python
# heatmap plot for getting correlation
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(housing_data_corr, annot=True)
```
