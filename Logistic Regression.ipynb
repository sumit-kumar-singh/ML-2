<b> LOGISTIC REGRESSION :<b> Logistic regression is supervised learning algorithm based on classification. It is mainly for the classification problems. It classifies output as binary value ( 0-no, 1-yes ) or may be multiple discrete value. A dataset with one or more independent variables is used to determine binary output of the dependent variable. In this we have sigmoid function or logistic function working behind the screen. A threshold is set ( p=0.5 ) and probability is determined for some event to occur. Applications are wheather predictions, image categorization, cancer types etc. Steps involved in number prediction:
    1. Importing libraries and their associated methods.
    2. Determining total number of images and labels.
    3. Represtation of images and labels.
    4. Dividing data set into train and test data set.
    5. Create logistic regression model and train our data.
    6. Predict the output.
    7. Test and evaluate the model
    

```python
# importing libraries and their associated methods

# importing pandas library for representing data in tabular form (dataframes)
import pandas as pd

# importing numpy library for calulations and computational work (scientific calculation)
import numpy as np

# importing matplotlib and seaborn for the graphical represtation of dataset
import matplotlib.pyplot as plt
import seaborn as sbn

# importing these for testing our model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

# importing train_test_split a helper function to split allowed inputs into random train and test subsets 
from sklearn.model_selection import train_test_split

# importing LogisticRegression algorithm for our model
from sklearn.linear_model import LogisticRegression

# importing for load and return the digits dataset (classification), sklearn.utils.Bunch
from sklearn.datasets import load_digits
%matplotlib inline
digits = load_digits()
```

```python
# determining total number of images and labels
print('Image Data Shape : ', digits.data.shape)
print('Label Data Shape : ', digits.target.shape)
```

```python
# represtation of images and labels
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(digits.data[0:5], digits.target[0:5])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training : %i\n' % label, fontsize = 20)
```

```python
# splitting data set in training and test data sets
X = digits.data
y = digits.target
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2, test_size=.23)
```

```python
# getting shape of the data set
print('X_train : ', X_train.shape)
print('y_train : ', y_train.shape)
print('X_test : ', X_test.shape)
print('y_test : ', y_test.shape)
```

```python
# create logistic regression instance and train our data
logisticRegr = LogisticRegression()
logisticRegr.fit(X_train, y_train)
```

```python
# returns numpy array
# predict for one observation ( image )
print('Prediction for One Image : ', logisticRegr.predict(X_test[0].reshape(1, -1)))
print('Prediction for Multiple Images : ', logisticRegr.predict(X_test[0:10]))
print('Prediction for Entire Images : \n', logisticRegr.predict(X_test))
```

```python
# getting score
score = logisticRegr.score(X_test, y_test)
print('Score is : ', score)
print('Accuracy %  is : ', score*100)
```

```python
# creating confusion matrix
X_predictions = logisticRegr.predict(X_test)
cm = confusion_matrix(y_test, X_predictions)
print('Confusion Matrix is : \n', cm)
# most of data is concentrated diagonally
# more the number in diagonal better the accuracy is
```

```python
# plotting heat map for better visualization
plt.figure(figsize=(9,9))
sbn.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square=True, cmap='Blues_r');
plt.ylabel('Actual Label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0} %'.format(score*100)
plt.title(all_sample_title, size = 15);
```

```python
# testing the model
index = 0
classifiedIndex = []
for predict, actual in zip(X_predictions, y_test):
    if predict==actual:
        classifiedIndex.append(index)
    index +=1
plt.figure(figsize=(20, 3))
for plotIndex, wrong in enumerate(classifiedIndex[0:4]):
    plt.subplot(1, 4, plotIndex +1)
    plt.imshow(np.reshape(X_test[wrong], (8,8)), cmap=plt.cm.gray)
    plt.title("Predicted: {}, Actual: {}" .format(X_predictions[wrong], y_test[wrong]), fontsize=20)
```
