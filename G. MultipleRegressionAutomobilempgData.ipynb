```python
# importing libraries
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading processed data
automobile_df = pd.read_csv('datasets/auto-mgp-processed.csv')
automobile_df.sample(5)
```

```python
# creating data for training based on multiple feature
X = automobile_df[['displacement', 'horsepower', 'weight']]
Y = automobile_df['mpg']
```

```python
# spliting data into test and train data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# training and fitting data
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression(normalize=True).fit(x_train, y_train)
```

```python
# printing training score
print('Training Score :', linear_model.score(x_train, y_train))
```

```python
# predicting values
y_pred = linear_model.predict(x_test)
```

```python
# calculating r2 for model
from sklearn.metrics import r2_score
print('R Score :', r2_score(y_test, y_pred))
```

```python
# assigning columns name to predictors for making prediction
predictors = x_train.columns
predictors
```

```python
# getting coefficient of the data predictors
predictors_coef = pd.Series(linear_model.coef_, predictors).sort_values()
print(predictors_coef)
```

```python
# converting into dataframe
df_actual_predict = pd.DataFrame({'predicted':y_pred, 'actual':y_test})
df_actual_predict.head(10)
```

```python
# visualizing data
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(y_test, y_pred)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# line map
plt.figure(figsize=(20,10))
plt.plot(y_pred, label='Predicted')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Mpg')
plt.legend()
plt.show()
```
